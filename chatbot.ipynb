```python
"""
Simple Student Guidance Chatbot - Beginner Friendly Version
This is a simplified version of the chatbot, easy to understand and modify.
"""

import os
from datetime import datetime
from google import genai

# Your API key - same one from main.py
API_KEY = "AIzaSyCXqUkL0fz_x_1TgmTYKRvd2SCHf2KIoIw"

# Initialize the client
client = genai.Client(api_key=API_KEY)

# System message that tells the AI what to do
SYSTEM_MESSAGE = """
You are a Student Guidance Counselor. Help students with:
- Study tips and academics
- Career planning
- College selection
- Personal development
- Stress management
- Extracurricular activities

IMPORTANT: Only answer student-related questions. If the question is not about student guidance, say:
"I am not made or customized for this type of question. I'm specifically designed to help with student guidance topics only."
"""


def create_chat():
    """Create a new chat session with the AI."""
    chat = client.chats.create(
        model='gemini-2.5-flash',
        config={
            'system_instruction': SYSTEM_MESSAGE,
            'temperature': 0.7,
        }
    )
    return chat


def get_ai_response(chat, user_message):
    """Send a message to the AI and get a response."""
    try:
        response = chat.send_message(user_message)
        return response.text
    except Exception as e:
        return f"Error: {e}"


def save_chat(messages, filename="chat_log.txt"):
    """Save the conversation to a text file."""
    try:
        with open(filename, 'w', encoding='utf-8') as file:
            for msg in messages:
                file.write(f"[{msg['time']}]\n")
                file.write(f"You: {msg['user']}\n")
                file.write(f"AI: {msg['bot']}\n")
                file.write("-" * 70 + "\n\n")
        print(f"Chat saved to {filename}")
    except Exception as e:
        print(f"Error saving: {e}")


def main():
    """Main function to run the chatbot."""
    print("\n" + "=" * 70)
    print("STUDENT GUIDANCE CHATBOT (Simple Version)".center(70))
    print("=" * 70)
    print("\nHello! I'm your AI Student Guidance Counselor.")
    print("Ask me anything about studying, career, college, or personal growth!")
    print("Type 'quit' to exit or 'save' to save the conversation.\n")

    # Create chat session
    chat = create_chat()

    # List to store all messages
    messages = []

    # Chat loop
    while True:
        # Get user input
        user_input = input("You: ").strip()

        # Check if user wants to quit
        if user_input.lower() == 'quit':
            print("\nThank you for chatting! Goodbye!")
            # Ask if user wants to save
            save_choice = input("Save conversation? (yes/no): ").strip().lower()
            if save_choice in ['yes', 'y']:
                save_chat(messages)
            break

        # Check if user wants to save
        if user_input.lower() == 'save':
            save_chat(messages)
            continue

        # Skip empty messages
        if not user_input:
            print("Please type something!\n")
            continue

        # Get response from AI
        print("\nAI: ", end="")
        response = get_ai_response(chat, user_input)
        print(response)
        print("\n" + "-" * 70 + "\n")

        # Store the message
        messages.append({
            'time': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
            'user': user_input,
            'bot': response
        })


if __name__ == "__main__":
    main()
```
